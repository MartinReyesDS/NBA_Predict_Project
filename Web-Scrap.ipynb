```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# NBA team we will be analyzing
team = "PHI"
# NBA season we will be analyzing
year = "2014"
# basic or advanced stats
log_adv = "gamelog"
# URL page we will scraping (see image above)
url = "https://www.basketball-reference.com/teams/{}/{}/{}/".format(team, year, log_adv)
# this is the HTML from the given URL
html = urlopen(url)
soup = BeautifulSoup(html)
```

```python
# use findALL() to get the column headers
soup.findAll('thead', limit=2)
# use getText()to extract the text we need into a list
headers = [th.getText() for th in soup.findAll('thead', limit=2)[0].findAll('th')]
# exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
headers = headers[7:]
```

```python
# avoid the first header row
rows = soup.findAll('tr')[2:]
```

```python
game_logs = [[td.getText() for td in rows[i].findAll('td')] for i in range(len(rows))]
```

```python
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

```python
game_log_df = pd.DataFrame(game_logs, columns = headers)
game_log_df
```

```python
game_log_df.rename(columns={'\xa0':'Home/Away', 'Tm':'Pts', 'Opp': ''}, inplace=True)
```

```python
game_log_df = game_log_df.iloc[:,:23]
```

```python
game_log_df['Home'] = game_log_df['Home/Away'].apply(lambda x: 0 if (x == '@') else 1)
game_log_df['Team'] = team
game_log_df
```
